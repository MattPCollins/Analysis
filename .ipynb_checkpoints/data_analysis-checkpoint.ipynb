```python
# sample by sample analysis 
df = pd.read_csv('fact_streak_cleansed.csv')
df_test=out[df.groupby(['index_word'])['index_word'].count()>20]
# max(list(df.groupby(['index_word'])['index_word'].count()))
ind_list = list(df_test['word index'])
df_sample = df[df['index_word'].isin(ind_list)]
for i in ind_list:
    df_sample = df[df['index_word']==i]
    plt.figure()
#     sns.lineplot(data=df_sample, x='date', y='streak',marker="o").set_title('word {}'.format(i))
    plt.plot_date(df_sample['date'],df_sample['streak'],linestyle='solid')
    plt.xticks(df_sample['date'], df_sample['date'], rotation='vertical')

```

```python
# min days at each streak vs actual vs diff
# get max diff
intervals_general = [0, 1, 2, 4, 7, 12, 31]
intervals3 = [0, 1, 3, 7, 14, 31, 31]
intervals4 = [0, 1, 2, 3, 5, 10, 31]
intervals5 = [0, 1, 1, 2, 4, 6, 31]
intervals = [intervals_general,intervals3,intervals4,intervals5]
intervals_desc = ['intervals_general','intervals3','intervals4','intervals5']

def sum_n(lst,n):
    return sum(lst[:n])

for i,j in zip(intervals, intervals_desc):
    days_to_5 = 1 + sum_n(i, 5)
    print('{} days to reach streak 5 for sample "{}"'.format(days_to_5, j))
```

```python
import pandas as pd


fact_streak = '/Users/mattcollins/Projects/MyProject/KanjiProgram/csvs/fact_streak.csv'
df_fs = pd.read_csv(fact_streak)

def new_df(my_index):
    df2 = df_fs[df_fs['index_word'] == my_index].copy()
    df2['streak_prev'] = df2['streak'].shift(1)
    return df2
def find_decrease(df2, d):
    for index,row in df2.iterrows():
        if row['streak'] <= row['streak_prev']:
#             print(row['streak_prev'])
            newrow = row['index_word']
            d.append([newrow,row['streak_prev']])
    return d
d = []
indexes = df_fs.index_word.unique()

for i in indexes:
    if len(df_fs[df_fs['index_word'] == i]) > 8:
        df2 = new_df(i)
        d = find_decrease(df2, d)
    
df_new = pd.DataFrame (d,columns=['index_word','streak'])
grouped = df_new.groupby('streak').size()

grouped.plot(kind = 'bar')
```
